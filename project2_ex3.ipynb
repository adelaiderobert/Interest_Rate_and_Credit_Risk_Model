```python
import numpy as np
from scipy.stats import ncx2
import matplotlib.pyplot as plt
```

```python
r = 0.05
sigma = 0.4
s0 = 100
T = 1
K = np.array([60, 80, 100, 120, 140])
```

```python
CT = 0.25 * (np.exp(r*T)-1)* sigma **2 / r 

lam = np.exp(r * T)* s0 / CT
```

```python
F1 = ncx2.sf(K / CT,2, lam)
F2 = ncx2.sf(K / CT, 1e-12 , lam)

C0 = s0 * F1 - np.exp(-r*T) * K * F2 
C0 = np.round(C0, 5)
```

```python
fig, ax1 = plt.subplots(1,1, figsize = (8 ,4))

ax1.plot(K, C0, color = 'blue', marker = 'o')
ax1.set_title(f"European call option with maturity T = {T}, initial price  = {s0}")
ax1.set_xlabel("Strike")
ax1.set_ylabel("Price")
plt.grid()
plt.tight_layout()

```
